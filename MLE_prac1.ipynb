```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
model = LinearRegression()

weights = np.array([1,2,3,4,5]).reshape(-1,1)
prices = np.array([30,60,90,120,150])

model.fit(weights, prices)

plt.scatter(weights, prices, color='blue', label='Data Points')

plt.plot(weights, model.predict(weights), color='orange', label='Regression Line')

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title("Predict price of banana based on weight")
plt.xlabel("Weight in kg")
plt.ylabel("Price in Rupees")
plt.tight_layout()
plt.show()
```

# Data Pre processing

## Load a Dataset using Pandas

```python
import pandas as pd
data = pd.read_csv("student_scores.csv")
data.head(10)
```

```python
print('Columns:', data.columns.tolist())
print('\nInfo')
print(data.info())

print('\nDescribe')

# Gives statistical data
display(data.describe())
```

## Handling missing values

```python
# checks if null values are there.
print(data.isnull().sum())

# Fill missing values with mean
data['score'] = data['score'].fillna(data['score'].mean())

# Drop rows with missing values
data = data.dropna()
```
